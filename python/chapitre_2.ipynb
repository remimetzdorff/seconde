```python
%pylab qt
%matplotlib inline

import fitting
import pandas
import random as rd

def mysavefig(name):
    plt.tight_layout()
    plt.savefig("../images/"+name, dpi=200)
```

```python
filename = "densite_saumure.xlsx"

dfs = pandas.read_excel(filename, sheet_name="densite_saumure")
c = np.array(dfs["Concentration"])[1:]
d = np.array(dfs["Densite"])[1:]

plt.figure(figsize=(6,4))
ax = plt.subplot2grid((1,1), (0,0))
ax.plot(c, d)

ax.set_xlabel(r"Concentration massique $C_\mathrm{m}$ (g/L)")
ax.set_ylabel(r"Masse volumique $\rho$ (kg/L)")
ax.set_title("Ã‰volution de la masse volumique\nen fonction de la concentration en sel")
plt.grid('on', alpha=0.5)

mysavefig("densite_saumure.png")

uy = np.ones(len(c))*1
fitter = fitting.Fit("linear", x=c, y=d)
fit_params, ufit_params = fitter.fit(manualguess_params=None, verbosemode=True)
print(fit_params)
yfit = fitter.func(c, *fit_params)
plt.plot(c, yfit)
```

```python
n = 200

becher     = np.random.normal(loc=10, scale=0.5, size=n)
eprouvette = np.random.normal(loc=10, scale=0.1, size=n)
pipette    = np.random.normal(loc=10, scale=0.01, size=n)
```

```python
plt.figure(figsize=(6,2.5))
sps = (1,3)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (0,1))
ax3 = plt.subplot2grid(sps, (0,2))

for data, ax, color in zip([eprouvette, becher, pipette], [ax1, ax2, ax3], ["C0","C1","C2"]):
    ax.hist(data, bins=10, rwidth=0.9, color=color)
    ax.set_ylim(0,55)
    if ax != ax1:
        ax.set_yticklabels([])
    ax.set_xlabel("Masse (g)")

ax1.set_ylabel("Nombre de mesures")
ax1.set_title("Verrerie A")
ax2.set_title("Verrerie B")
ax3.set_title("Verrerie C")
mysavefig("precision_verrerie.png")
```
