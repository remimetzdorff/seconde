```python
%pylab qt
%matplotlib inline
import pandas

def mysavefig(name):
    plt.tight_layout()
    plt.savefig("../images/"+name, dpi=200)
```

```python
def make_series(X,Y):
    listx, listy = [], []
    for x,y in zip(X,Y):
        listx += list(x)
        listy += list(y)
    return pandas.Series(listy, index=listx)

t1 = np.linspace(0,3,101)
t2 = np.linspace(3,7,101)
t3 = np.linspace(7,10,101)

#eau pure
y1 = 2 * (t1-3)
y2 = np.zeros(len(t2))
y3 = 2 * (t3-7)
data_pur     = make_series([t1,t2,t3], [y1,y2,y3])

#eau salee
y1 = 2.5 * (t1-3) - 3
y2 = 0.3 * (t2-3) - 3
y3 = 2.5 * (t3-7) - 1.8
data_melange = make_series([t1,t2,t3], [y1,y2,y3])

plt.figure(figsize=(4,3))

plt.plot(data_pur, label="eau pure")
plt.plot(data_melange, label="eau salée")

plt.xlabel("Temps (min)")
plt.ylabel("Température (°C)")
plt.legend(loc="upper left")

mysavefig("fusion.png")
```

```python
data = np.loadtxt("densite_eau_sucree.txt", skiprows=2, dtype=np.str, delimiter='\t')
data = np.char.replace(data, ',', '.').astype(np.float64)

brix    = data[:,0]
density = data[:,1]

plt.figure(figsize=(6,4))
ax = plt.subplot2grid((1,1), (0,0))
plt.plot(brix, density)
plt.xlabel("Pourcentage massique de saccharose dans le sirop (%)")
plt.ylabel("Masse volumique du sirop (kg/L)")
plt.grid('on')
ax.set_yticks([1.00, 1.05, 1.10, 1.15, 1.20, 1.25, 1.30, 1.35, 1.40])

mysavefig("densite_sirop.png")
```
